# Ten most popular songs

```python
most_popular=df_tracks[df_tracks['popularity']>90].sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
df_tracks.set_index('release_date',inplace=True)
df_tracks.index = df_tracks.index.map(pd.to_datetime)
df_tracks.head()
```

```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

# Correlation Matrix

```python
corr_df=df_tracks.drop(df_tracks.select_dtypes(include='object'), axis=1).corr(method='pearson')
plt.figure(figsize=(16,8))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',annot_kws={'fontsize':10,'color':'Black'},vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
heatmap.set_title('Correlation HeatMap between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='energy', y='loudness', color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='acousticness', y='popularity', color='b').set(title='Popularity vs Acousticness Correlation')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.distplot(years).set(title='Number of songs per year')
```

```python
total_dr=df_tracks.duration
fig,ax=plt.subplots(figsize=(10,7))
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title='Year vs Duration')
plt.xticks(rotation=90)
```

```python
sns.set_style(style='whitegrid')
fig,ax=plt.subplots(figsize=(10,5))
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title='Year vs Duration')
plt.xticks(rotation=60)
```

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

```python
df_genre.info()
```

```python
df_genre.isna().sum()
```

```python
df_genre.describe().transpose()
```

```python
genre_grouped=df_genre.groupby('genre')['duration_ms'].mean()
genre_grouped
```

```python
plt.title('Duration of the Songs in Different Genres')
sns.color_palette('rocket',as_cmap=True)
sns.barplot(y=genre_grouped.index,x=genre_grouped.values)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

# Top 5 Genres by Popularity

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous=df_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title='Top 5 Genres by Popularity')
```

```python
corr_df=df_genre.drop(['genre','artist_name','track_name','mode','track_id','key','time_signature'],axis=1).corr(method='pearson')
plt.figure(figsize=(16,8))
heatmap=sns.heatmap(corr_df,annot=True,cmap='inferno',linewidths=1,linecolor='Black')
heatmap.set_title('Correlation HeatMap between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x='energy',y='loudness',data=df_genre,line_kws={'color':'b'},color='c').set(title='Energy vs Loudness')
plt.show()
```
